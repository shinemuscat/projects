```python
df['리뷰']=df['리뷰'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['리뷰']=df['리뷰'].str.replace('  ','')
df['리뷰']=df['리뷰'].str.replace('ㅋ','')
df['리뷰']=df['리뷰'].str.replace('ㅎ','')
df['리뷰']=df['리뷰'].str.replace('ㅜ','')
df['리뷰']=df['리뷰'].str.replace('ㅠ','')
df['리뷰']
```

```python
with open('불용어리스트.txt', "r", encoding='UTF-8') as file:
    strings = file.readlines()
    #print(strings)
```

```python
불용어=pd.DataFrame({'불용어':strings})
불용어=불용어['불용어'].str.replace('\n','')
stopwords=불용어.to_list()
```

```python
from eunjeon import Mecab
m = Mecab()
```

```python
sent='안전성에 문제있는 스마트폰을 휴대하고 탑승 할 경우에 압수한다'
m.morphs(sent)
```

```python

tokenized_data = []
for sentence in df['리뷰']:
    temp_X = m.morphs(sentence) # 토큰화
    temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
    tokenized_data.append(temp_X)
#     print('불용어=>',temp_X)
```

```python
len(tokenized_data)
```

```python
print(tokenized_data[0])
```

```python
import numpy as np
words = np.hstack(tokenized_data)
```

```python
words
```

```python
from collections import Counter

words_counter = Counter(words)
words_counter = words_counter.most_common()
```

```python
# words_counter
```

```python
words_counter_pos = []
for item in words_counter:
  pos_tuple = (item[0], item[1], m.pos(item[0]))
  words_counter_pos.append(pos_tuple)
```

```python
words_counter_pos[:5]
```

```python
wcp_df = pd.DataFrame(words_counter_pos)
wcp_df.columns = ['word', 'count', 'pos']
wcp_df.head(30)
```
