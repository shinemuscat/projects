```python
import os
import re
import string
```

```python
base_dir = 'C://Users//ILIFO-001//NER-master//말뭉치 - 형태소_개체명'
file_list = os.listdir(base_dir)
```

```python
import chardet
def get_encoding(path):
    char_dic = chardet.detect(open(path, 'rb').read())
    return char_dic['encoding']
```

```python
num = 0
cnt = 1
for file in file_list:
    file_name, file_ext = os.path.splitext(file)
    if (file_name != 'EXO') & (file_ext == '.txt'):
        encoding = get_encoding(os.path.join(base_dir, file))
        with open(os.path.join(base_dir, '{}.txt'.format(file_name)), 'r', encoding=encoding) as fr, \
             open(os.path.join(base_dir, '{}.csv'.format(file)), 'w', encoding=encoding) as fw :
            for line in fr:
                if re.search('## [0-9]+', line):
                    match = re.search('## ([0-9]+)', line)            
                    num = match.group(1)
                if re.search('^[a-zA-Z0-9\u3131-\u3163\uac00-\ud7a3]+', line):
                    fw.write('sentence_{}_{}\t'.format(file_name, num) + line)
                    cnt += 1
```

```python
base_dir = 'C://Users//ILIFO-001//NER-master//말뭉치 - 형태소_개체명'
file_list = os.listdir(base_dir)
```

```python
import pandas as pd

df_list = []
for file in file_list:
    file_name, file_ext = os.path.splitext(file)    
    if file_ext == '.csv':
        encoding = get_encoding(os.path.join(base_dir, file))
        df_list.append(pd.read_csv(os.path.join(base_dir, '{}.csv'.format(file_name)), header=None, sep='\t', \
                        names=['Sentence #', 'Word','word_pos', 'POS', 'Tag'], encoding=encoding))
df = pd.concat(df_list)  
df.reset_index(drop = True, inplace = True)
print(len(df))
```
